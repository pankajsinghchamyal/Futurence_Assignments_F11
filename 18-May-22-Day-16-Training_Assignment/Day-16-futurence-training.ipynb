```python
# make the data
np.random.seed(3)
x = np.array([1,2,4,6])
y = np.array([4,5,6,8])
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,5):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
sinplot()
plt.show()
```

```python
def rec(start,ans):
    if start>=10:
        return ans
    else:
        start+=1
        ans+=start
        return rec(start,ans)
print(rec(1,0))
```
